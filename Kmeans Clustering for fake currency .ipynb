```python
#Importing and reading csv file for data

import pandas as pd

data = pd.read_csv('Banknote-authentication-dataset.csv')

V1 = data['V1']
V2 = data['V2']

V1.min()
```

```python
import numpy as np


mean_V1 = np.mean(V1)
mean_V2 = np.mean(V2)

std_V1 = np.std(V1)
std_V2 = np.std(V2)

print(mean_V1, mean_V2)
print(std_V1, std_V2)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(V1,V2,s=50,alpha=0.5)
```

```python
from sklearn.cluster import KMeans

stack = np.column_stack((V1,V2))
km_res = KMeans(n_clusters = 2).fit(stack)
clusters = km_res.cluster_centers_
labels = km_res.labels_

plt.xlabel('V1')
plt.ylabel('V2')

plt.scatter(V1,V2,s = 50,c = labels, alpha = 0.6)
plt.scatter(clusters[:,0], clusters[:,1], s=1500, alpha=0.8 ,c='red')
# plt.savefig("fig1")
```
