```python
import pandas as pd

data = pd.read_csv('happyscore_income.csv')
```

```python
happy = data['happyScore']
income = data['avg_income']

income.min()
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Income')
plt.ylabel('Happy score')
plt.scatter(income, happy)
```

```python
# Better Visualization Using more than 2 dimensions 

happy = data['happyScore']
ineq = data['income_inequality']
income = data['avg_income']

plt.scatter(income, happy, s=ineq*10, alpha= 0.25)
```

```python
# K Clustering using KMeans

from sklearn.cluster import KMeans
import numpy as np

happy = data['happyScore']
income = data['avg_income']
income_happy = np.column_stack((income,happy))

richest = data[ data['avg_income'] > 10000 ]
poorest = data[ data['avg_income'] < 10000 ] 

km_res = KMeans(n_clusters = 3).fit(income_happy)
clusters = km_res.cluster_centers_

plt.xlabel("Avg Income")
plt.ylabel("Happiness")

plt.text(richest.iloc[-1]['avg_income'],
         richest.iloc[-1]['happyScore'],
         richest.iloc[-1]['country'])

plt.text(poorest.iloc[0]['avg_income'],
         poorest.iloc[0]['happyScore'],
         poorest.iloc[0]['country'])

plt.text(poorest.iloc[-1]['avg_income'],
         poorest.iloc[-1]['happyScore'],
         poorest.iloc[-1]['country'])

plt.scatter(income,happy,alpha=0.5)
plt.scatter(clusters[:,0], clusters[:,1], s=1000, alpha=0.8)
```
